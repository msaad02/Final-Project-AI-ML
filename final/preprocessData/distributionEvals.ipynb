```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv("../../data/kaggleDataset/chessData.csv")
```

```python
def preprocess_scores(eval):
    score_dict = {
        '#+': 30000,
        '#-': -30000
    }

    try:
        eval = int(eval)
        eval = max(min(eval, 15000), -15000)
    except ValueError:
        eval = score_dict.get(eval[0:2], 0)

    return eval
```

```python
import matplotlib.pyplot as plt
# data = df.Evaluation.apply(preprocess_scores)  # arguments are passed to np.histogram

plt.hist(df.Evaluation, bins=100, range=(-5000, 5000), log=True)
plt.title("Histogram with 'auto' bins")
plt.show()
```

```python
import matplotlib.pyplot as plt
data = df.Evaluation.apply(preprocess_scores)  # arguments are passed to np.histogram

plt.hist(data, bins=100, range=(-2000, 2000), log=False)
plt.title("Histogram with 'auto' bins")
plt.show()
```
